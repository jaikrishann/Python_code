```python
sns.relplot(x='TopSpeed_KmH',y='Range_Km',data=electric_car,kind='scatter')
```

```python
sns.relplot(x='TopSpeed_KmH',y='Range_Km',data=electric_car,kind='line')
```

```python
sns.relplot(x='TopSpeed_KmH',y='Range_Km',data=electric_car,kind='scatter',hue='PriceEuro')
```

```python
sns.relplot(x='TopSpeed_KmH',y='Range_Km',data=electric_car,kind='scatter',hue='PriceEuro',style='RapidCharge')
```

```python
sns.relplot(x='TopSpeed_KmH',y='Range_Km',data=electric_car,kind='scatter',hue='PriceEuro',style='RapidCharge',col='RapidCharge')
```

```python
sns.catplot(x='BodyStyle',y='AccelSec',data=electric_car,kind='strip')
```
